```python
df["Edad"].hist(bins=20)
plt.xlabel("edad")
plt.ylabel("count")
plt.show()
```

```python
estado_count = df["Atención**"].value_counts()
plt.bar(estado_count.index,estado_count.values)
plt.xticks([0,1,2,3,4],["Recuperado","Casa","Hospital","Hospital UCI","Fallecido"])
plt.ylabel("count")
plt.show()
```

```python
df.Sexo.value_counts().plot.pie()
plt.show
```

```python
departamento = df.groupby("Departamento o Distrito")["Edad"].mean()
departamento.head(14).plot.barh()
plt.show()
```
